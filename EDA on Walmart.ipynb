# EDA on walmart Product Dataset

Our Dataset contains the product details such as their Description, url,  Price in USD, brand, main image, SKU, primary category, Sub_Category_1, Sub_category_2, images, Specifications, Directions to use the product, Ingredients, Average rating, Reviews count. I highly recommend you to take a look at the dataset yourself to understand it better.
[link to dataset](https://www.kaggle.com/datasets/hikageshinomori/walmart-e-commerce-product-data?select=walmart.csv)

```python
#Previewing our dataset of walmart product data using pandas
import pandas as pd

df = pd.read_csv('/content/walmart.csv')
print(df.head())
```

```python
#Getting a concise summary of our data
print(df.info(verbose=True))
```

```python
#Identifying columns with missing or null values
df.isnull().sum()
```

```python
#getting a statistical summary of numerical columns
df.describe()
```

```python
#Data cleaning (removing duplicates)
df.drop_duplicates()
```

```python
#creating this pivot table to find the number of unique products in each primary category
pivot_table = df.pivot_table(index=['primary_category'],values=['title'],aggfunc='count')
print(pivot_table)
```

Created a pivot table for primary category to title to find the number of total products in each primary category

```python
#plotting the number of products in each primary_category using matplotlib.pyplot
import matplotlib.pyplot as plt
pivot_table.plot(kind='bar')
plt.show()
```

```python
#Creating the pivot table for reviews
pivot_table2 = df.pivot_table(index=['primary_category'],values=['reviews_count'],aggfunc='sum')
print(pivot_table2)
```

Created this pivot table to find the number of reviews on each primary category of products to get highly reviewed categories.

```python
#And now plotting the reviews pivot table
pivot_table2.plot(kind='bar')
plt.show()
```

We can further assume that the primary category of Household Essentials has the most number of products present in the dataset, and we can further say that the personal care products are the most reviewed products.
